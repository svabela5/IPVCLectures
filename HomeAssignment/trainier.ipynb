```python
from ultralytics import YOLO
import numpy as np
```

```python
model = YOLO('yolo11n.pt')
```

```python
results = model.train(
    data='finalDataset.yaml',
    epochs=100,
    imgsz=640,
    batch=16,
    name='FinalAIModel',
    patience=10,
    project='AI Models'
)
```

```python
model = YOLO('AI Models/FinalAIModel/weights/best.pt')
print("--- STARTING DIAGNOSTIC RUN ---")
metrics = model.val(data='finalDataset.yaml', split='val')

print("\n--- DIAGNOSTIC RESULTS ---")
print(f"Mean Precision: {metrics.box.mp:.4f}")
print(f"Mean Recall: {metrics.box.mr:.4f}")
print(f"mAP@50: {metrics.box.map50:.4f}")
print(f"mAP@50-95: {metrics.box.map:.4f}")

# CHECK 4: Manual F1 Calculation (Fallback)
mp, mr, map50, map95 = metrics.box.mean_results()
if (mp + mr) > 0:
    f1 = 2 * (mp * mr) / (mp + mr)
    print(f"Calculated F1 Score: {f1:.4f}")
else:
    print("Calculated F1 Score: 0.0 (Model has 0 Precision or 0 Recall)")
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
imgBGR= cv2.imread('images/lena.jpg')
imgGray=cv2.cvtColor(imgBGR, cv2.COLOR_BGR2GRAY)
```

```python
#Refer to : https://opencv.org/blog/edge-detection-using-opencv/

img_gaussian = cv2.GaussianBlur(imgGray, (3,3),0 )

# Apply Laplacian operator
laplacian = cv2.Laplacian(img_gaussian, cv2.CV_64F)
 
# Convert to uint8
laplacian_abs = cv2.convertScaleAbs(laplacian)

sobelx= cv2.Sobel(img_gaussian, cv2.CV_64F, 1, 0, ksize=5)
sobely= cv2.Sobel(img_gaussian, cv2.CV_64F, 0, 1, ksize=5)

plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
plt.imshow(img_gaussian, cmap='gray', vmin=0, vmax=255)
plt.title('Original')

plt.subplot(2,2,2)
plt.imshow(laplacian_abs, cmap='gray', vmin=0, vmax=255)
plt.title('Laplacian')

plt.subplot(2,2,3)
plt.imshow(sobelx, cmap='gray', vmin=0, vmax=255)
plt.title('Sobel X')

plt.subplot(2,2,4)
plt.imshow(sobely, cmap='gray', vmin=0, vmax=255)
plt.title('Sobel Y')
plt.show()

sobelXY = sobelx + sobely
plt.imshow(sobelXY, cmap='gray', vmin=0, vmax=255)
plt.title('Sobel XY')
plt.show()
```

```python
#CANNY EDGE DETECTOR

imgBGR= cv2.imread('images/lena.jpg')
imgGray=cv2.cvtColor(imgBGR, cv2.COLOR_BGR2GRAY)

img_gaussian = cv2.GaussianBlur(imgGray, (3,3),0 )

imgCanny = cv2.Canny(img_gaussian, 40, 100)
plt.imshow(imgCanny, cmap='gray', vmin=0, vmax=255)
plt.show()


```

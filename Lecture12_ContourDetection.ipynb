```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
imBGR = cv2.imread('images/thumbs_up_down.jpg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
imgGrey = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)
```

```python
#create binary image

_, binary = cv2.threshold(imgGrey, 220, 255, cv2.THRESH_BINARY_INV)
plt.imshow(binary, cmap='gray', vmin=0, vmax=255)
plt.show()

#find contours
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(contours)
print(f"Number of countours found: {len(contours)}")

output = cv2.drawContours(imRGB.copy(), contours, -1, (0,255,0), 2)
plt.imshow(output)
```

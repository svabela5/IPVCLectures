```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
imBGR = cv2.imread('images/thumbs_up_down.jpg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
imgGrey = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)
```

```python
#create binary image

_, binary = cv2.threshold(imgGrey, 220, 255, cv2.THRESH_BINARY_INV)
plt.imshow(binary, cmap='gray', vmin=0, vmax=255)
plt.show()

#find contours
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(contours)
print(f"Number of countours found: {len(contours)}")

output = cv2.drawContours(imRGB.copy(), contours, -1, (0,255,0), 2)
plt.imshow(output)
```

```python
#create binary image

imBGR = cv2.imread('images/noise_out_thumbs_up_down.jpg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
imgGrey = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)

_, binary = cv2.threshold(imgGrey, 220, 255, cv2.THRESH_BINARY_INV)
plt.imshow(binary, cmap='gray', vmin=0, vmax=255)
plt.title('Binary Image with Noise')
plt.show()

# Perform Morphological opening to remove noise
SEKernel = np.ones((5,5), np.uint8)
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, SEKernel)
plt.title('After Morphological Opening')
plt.imshow(opening, cmap='gray', vmin=0, vmax=255)
plt.show()

#find contours
contours, hierarchy = cv2.findContours(opening, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(contours)
print(f"Number of countours found: {len(contours)}")

output = cv2.drawContours(imRGB.copy(), contours, -1, (0,255,0), 2)
plt.imshow(output)
```

```python
#Iteratig through the list of contours to perform countour operations
imBGR = cv2.imread('images/noise_out_thumbs_up_down.jpg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
imgGrey = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)

_, binary = cv2.threshold(imgGrey, 220, 255, cv2.THRESH_BINARY_INV)
plt.imshow(binary, cmap='gray', vmin=0, vmax=255)
plt.title('Binary Image with Noise')
plt.show()

# Perform Morphological opening to remove noise
SEKernel = np.ones((5,5), np.uint8)
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, SEKernel)
plt.title('After Morphological Opening')
plt.imshow(opening, cmap='gray', vmin=0, vmax=255)
plt.show()

#find contours
contours, hierarchy = cv2.findContours(opening, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
print(contours)
print(f"Number of countours found: {len(contours)}")

for c in contours:
    hull = cv2.convexHull(c)
    cv2.drawContours(imRGB, [hull], 0, (255,0,0), 2)
    cv2.drawContours(imRGB, [c], 0, (0,255,0), 2)

# output = cv2.drawContours(imRGB.copy(), contours, -1, (0,255,0), 2)
plt.imshow(imRGB)
```

```python
#Iteratig through the list of contours to perform countour operations
imBGR = cv2.imread('images/noise_out_thumbs_up_down.jpg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
imgGrey = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)

_, binary = cv2.threshold(imgGrey, 220, 255, cv2.THRESH_BINARY_INV)
plt.imshow(binary, cmap='gray', vmin=0, vmax=255)
plt.title('Binary Image with Noise')
plt.show()

# Perform Morphological opening to remove noise
SEKernel = np.ones((5,5), np.uint8)
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, SEKernel)
plt.title('After Morphological Opening')
plt.imshow(opening, cmap='gray', vmin=0, vmax=255)
plt.show()

#find contours
contours, hierarchy = cv2.findContours(opening, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
# print(contours)
print(f"Number of countours found: {len(contours)}")

count = 0
for c in contours:
    count += 1
    #Centroid
    M = cv2.moments(c)
    cX = int(M['m10'] / M['m00'])
    cY = int(M['m01'] / M['m00'])
    cv2.circle(imRGB, (cX, cY), 5, (0,0,255), -1)
    cv2.putText(imRGB, f"{count}", (cX - 50, cY - 50), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 2)

    area = cv2.contourArea(c)
    print(f"Contour {count} has area: {area}")
    perimiter = cv2.arcLength(c, True)
    print(f"Contour {count} has perimiter: {perimiter}")    

    cv2.drawContours(imRGB, [c], 0, (0,255,0), 2)

# output = cv2.drawContours(imRGB.copy(), contours, -1, (0,255,0), 2)
plt.imshow(imRGB)
```

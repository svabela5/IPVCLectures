```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
imBGR = cv2.imread('images/lena.jpg')

imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
plt.imshow(imRGB)
plt.show()

im_R = imRGB.copy()
im_R[:, :, (1, 2)] = 0
plt.imshow(im_R)
plt.show()

# Obtain and display cyan channel of the image
im_C = 255 - im_R
plt.imshow(im_C)
plt.show()

# Obtain and display the Magenta channel only
im_G = imRGB.copy()
im_G[:, :, (0, 2)] = 0
im_M = 255-im_G
plt.imshow(im_M)
plt.show()

# Obtain and display the Yellow channel only
im_B = imRGB.copy()
im_B[:, :, (0, 1)] = 0
im_Y = 255 - im_B
plt.imshow(im_Y)
plt.show()
```

```python
# Using the HSV color space

imBGR = cv2.imread('images/lena.jpg')

imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
plt.imshow(imRGB)
plt.show()

imHSV = cv2.cvtColor(imBGR, cv2.COLOR_BGR2HSV)
plt.imshow(imHSV)
plt.show()
# note that we cannot and it doesn't make sense to display HSV images. HSV is used for processing

imHSV[:, :, 1] = 0  # set saturation to max
imOutRGB = cv2.cvtColor(imHSV, cv2.COLOR_HSV2RGB)
plt.imshow(imOutRGB)
plt.show()

imHSV[20:30, 45:55, 1] = 255  # set hue to 0 in a small patch
imOutRGB_2 = cv2.cvtColor(imHSV, cv2.COLOR_HSV2RGB)
plt.imshow(imOutRGB_2)
plt.show()
```

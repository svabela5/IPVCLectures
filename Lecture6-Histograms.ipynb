```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Example : Generate a random grayscale 2D matrix 10 x 10 pixels with the following possible pixel values: 50, 140, 160, 180, 190
# Perform histogram normalisation ad dispaly the output in a 1 by 2 subplot

possible_values =np.array([50, 140, 160, 180, 190], dtype='uint8')
randomimg = np.random.choice(possible_values, size=(10, 10))
print(randomimg)

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(randomimg, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 2, 2)
plt.title('Original Histogram')
hist = cv2.calcHist([randomimg], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()

normalized = cv2.normalize(randomimg, None, 0, 255, cv2.NORM_MINMAX)

plt.subplot(1, 2, 1)
plt.title('Normalized Image')
plt.imshow(normalized, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 2, 2)
plt.title('Normalized Histogram')
hist = cv2.calcHist([normalized], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

### Equalisation

```python
#Example : Generate a random grayscale 2D matrix 10 x 10 pixels with the following possible pixel values: 50, 140, 160, 180, 190
# Perform histogram Equalisation ad dispaly the output in a 1 by 2 subplot

possible_values =np.array([50, 140, 160, 180, 190], dtype='uint8')
randomimg = np.random.choice(possible_values, size=(10, 10))
print(randomimg)

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(randomimg, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 2, 2)
plt.title('Original Histogram')
hist = cv2.calcHist([randomimg], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()

eqaulized = cv2.equalizeHist(randomimg)

plt.subplot(1, 2, 1)
plt.title('Eqaulized Image')
plt.imshow(eqaulized, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 2, 2)
plt.title('Eqaulized Histogram')
hist = cv2.calcHist([eqaulized], [0], None, [256], [0, 256])
plt.plot(hist)
plt.show()
```

```python
M2 = np.random.randint(0, 256, size=M.shape, dtype='uint8')
plt.imshow(M2, cmap='gray', vmin=0, vmax=255)
```

```python
M3 = np.zeros([100,200,3], dtype='uint8')
M3[:,:100,0] = 255
M3[:,:100,1] = 128

M3[:,100:,2] = 255
plt.imshow(M3, cmap='gray', vmin=0, vmax=255)

save_output = Image.fromarray(M3)
save_output.save('Question6.bmp')
```

```python
M3 = np.zeros([100,200,4], dtype='uint8')
M3[:,:100,0] = 255
M3[:,:100,1] = 128

M3[:,100:,2] = 255
M3[:,:,3] = np.linspace(0,255,200,dtype='uint8')# creates gradint
plt.imshow(M3, cmap='gray', vmin=0, vmax=255)

# 0 = Transparent
# Other extreme is 255 which is opaque
```

```python
M3 = np.zeros([100,200,4], dtype='uint8')
M3[:,:100,0] = 255
M3[:,:100,1] = 128

M3[:,100:,2] = 255

H = 100
gradient = np.linspace(0, 255, H, dtype='uint8')
M3[:, :, 3] = gradient[:, np.newaxis]

plt.imshow(M3, cmap='gray', vmin=0, vmax=255)
# 0 = Transparent
# Other extreme is 255 which is opaque
```

```python
#Extra example
import numpy as np
import matplotlib.pyplot as plt

M4 = np.zeros((100, 200, 3), dtype="uint8")


for x in range(M4.shape[0]):
    M4[x, :, 0] = x  # Set transparency gradient from 0 to 255
plt.imshow(M4)
plt.show()
```

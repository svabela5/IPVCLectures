### Task 1

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
# 1. 
imBGR = cv2.imread('images/goldfish.jpg')
imGray = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)

# 2.
t_value = int(input("Enter threshold value (0-255): "))

# 3.
im_result = t_value - imGray

# 4.
imNewResult = np.clip(t_value - imGray, 0, 255).astype(np.uint8)

plt.imshow(imNewResult, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Task 2

```python
# 2.
imBGR = cv2.imread('images/goldfish.jpg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

fig, ax = plt.subplots(2, 2)

ax[0, 0].imshow(imRGB)
ax[0, 0].set_title('Original Image')
ax[0, 0].axis('off')

imr = imRGB.copy()
imr[:, :, 1] = 0 #Turn off green channels
imr[:, :, 2] = 0 #Turn off blue channels
ax[0, 1].imshow(imr)
ax[0, 1].set_title('Red Channel')
ax[0, 1].axis('off')

img = imRGB.copy()
img[:, :, 0] = 0 #Turn off Red channels
img[:, :, 2] = 0 #Turn off blue channels
ax[1, 0].imshow(img)
ax[1, 0].set_title('Green Channel')
ax[1, 0].axis('off')

imb = imRGB.copy()
imb[:, :, 1] = 0 #Turn off green channels
imb[:, :, 0] = 0 #Turn off Red channels
ax[1, 1].imshow(imb)
ax[1, 1].set_title('Blue Channel')
ax[1, 1].axis('off')

plt.plot()

outputs = {
    "w3t2n4_Red" : imr,
    "w3t2n4_Green" : img,
    "w3t2n4_Blue" : imb,
    "w3t2n4_Colors" : imRGB,
}

for key, value in outputs.items():
    saveOutput = Image.fromarray(value)
    saveOutput.save(f"{key}.png")
```

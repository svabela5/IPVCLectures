```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Task 1

```python
imBGR = cv2.imread('images/lena.jpg')

imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
plt.imshow(imRGB)
plt.show()

imHSV = cv2.cvtColor(imBGR, cv2.COLOR_BGR2HSV)
imHSV_Brightend = np.copy(imHSV)
imHSV_Brightend[:, :, 2] = np.clip(imHSV_Brightend[:, :, 2].astype(int) + 100, 0, 254).astype(np.uint8)

imRGB_Brightend = cv2.cvtColor(imHSV_Brightend, cv2.COLOR_HSV2RGB)
plt.imshow(imRGB_Brightend)
plt.show()
```

### Task 2

```python
imBGR = cv2.imread('images/lena.jpg')

imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
plt.imshow(imRGB)
plt.show()

brightness = int(input("Enter the brightness adjustment"))

imHSV = cv2.cvtColor(imBGR, cv2.COLOR_BGR2HSV)
imHSV_Brightend = np.copy(imHSV)
imHSV_Brightend[:, :, 2] = np.clip(imHSV_Brightend[:, :, 2].astype(int) + brightness, 0, 254).astype(np.uint8)

imRGB_Brightend = cv2.cvtColor(imHSV_Brightend, cv2.COLOR_HSV2RGB)
plt.imshow(imRGB_Brightend)
plt.show()
```

### Task 3

```python
imBGR = cv2.imread('images/bird2.jpg')
imHSV = cv2.cvtColor(imBGR, cv2.COLOR_BGR2HSV)

imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
plt.imshow(imRGB)
plt.show()

lowLowRed = np.array(np.array([0, 100, 50]))
lowHighRed = np.array([10, 255, 255])

highLowRed = np.array([170, 100, 50])
highHighRed = np.array([179, 255, 255])

lowMask = cv2.inRange(imHSV, lowLowRed, lowHighRed)
highMask = cv2.inRange(imHSV, highLowRed, highHighRed)

mergedMask = cv2.bitwise_or(lowMask, highMask)

Result_BGR = cv2.bitwise_and(imBGR,imBGR, mask=mergedMask)
Result_RGB = cv2.cvtColor(Result_BGR, cv2.COLOR_BGR2RGB)

plt.imshow(Result_RGB)
plt.plot()
```

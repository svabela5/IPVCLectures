```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Task 1

```python
imBGR = cv2.imread('images/lena.jpg')
imGray = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)
imrgb = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

plt.subplot(2, 2, 1)
plt.title('Gray Image')
plt.imshow(imGray, cmap='gray', vmin=0, vmax=255)

plt.subplot(2, 2, 2)
plt.title('Gray Histogram')
hist = cv2.calcHist([imGray], [0], None, [256], [0, 256])
plt.plot(hist)

plt.subplot(2, 2, 3)
plt.title('RGB Image')
plt.imshow(imrgb)

plt.subplot(2, 2, 4)
plt.title('RGB Histograms')
colors = ('r', 'g', 'b')
for i, color in enumerate(colors):
    hist = cv2.calcHist([imrgb], [i], None, [256], [0, 256])
    plt.plot(hist, color=color)
```

# Task 2

```python
imBGR = cv2.imread('images/lena.jpg')
imGray = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)

lena_equalized = cv2.equalizeHist(imGray)

lena_concatinated = np.hstack((imGray, lena_equalized))

plt.imshow(lena_concatinated, cmap='gray', vmin=0, vmax=255)
plt.title('Original and Equalized Image')
plt.show()

plt.subplot(1, 2, 1)
plt.title('Original histogram')
hist = cv2.calcHist([imGray], [0], None, [256], [0, 256])
plt.plot(hist)
plt.subplot(1, 2, 2)
plt.title('Equalized histogram')
hist_eq = cv2.calcHist([lena_equalized], [0], None, [256], [0, 256])
plt.plot(hist_eq)

plt.show()
```

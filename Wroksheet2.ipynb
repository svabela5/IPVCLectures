```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

```python
imBGR = cv2.imread('images/goldfish.jpg')

imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

redimRGB = imRGB.copy()
redimRGB[:, :, 1] = 0  # Set the green channel to zero
redimRGB[:, :, 2] = 0  # Set the blue channel to zero
plt.imshow(redimRGB)
plt.show()

greenimRGB = imRGB.copy()
greenimRGB[:, :, 0] = 0  # Set the Red channel to zero
greenimRGB[:, :, 2] = 0  # Set the blue channel to zero
plt.imshow(greenimRGB)
plt.show()

blueImRGB = imRGB.copy()
blueImRGB[:, :, 1] = 0  # Set the green channel to zero
blueImRGB[:, :, 0] = 0  # Set the red channel to zero
plt.imshow(blueImRGB)
plt.show()
```

```python
imBGR = cv2.imread('images/goldfish.jpg')

imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

addedRGB = np.clip(imRGB.astype(int) + 5, 0, 255).astype('uint8')
plt.imshow(addedRGB)
plt.show()
```

```python
imBGR = cv2.imread('images/goldfish.jpg')

imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

plt.imshow(imRGB[20:-20, 20:-20])
plt.show()
```

```python
imBGR = cv2.imread('images/goldfish.jpg')

imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

imRGBFlipped = imRGB[::-1, ::-1, :]
plt.imshow(imRGBFlipped)
plt.show()
```

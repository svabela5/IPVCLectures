```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.array(['Jan', 'Feb', 'Mar', 'Apr'])
y = np.array([45, 78, 32, 23])

plt.bar(x, y)
plt.xlabel('Months')
plt.ylabel('Rainfall in mm')
plt.title('Rainfall')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,2])

x = np.array(['Jan', 'Feb', 'Mar', 'Apr'])
y = np.array([45, 78, 32, 23])

ax.bar(x, y)
ax.set_xlabel('Months')
ax.set_ylabel('Rainfall in mm')
ax.set_title('Rainfall')
plt.show()
```

```python
a = np.array([92, 87, 5, 45, 9, 78, 4, 34, 29, 26, 92, 95, 98, 91, 56, 40, 63])
bins = [0, 25, 50, 75, 100]

plt.hist(a, bins)
plt.ylabel('No of Students')
plt.xlabel('Grades')
plt.title('Exam Reuslts')
```

How to fix issue with x axis labeling that can be seen above

```python
a = np.array([92, 87, 5, 45, 9, 78, 4, 34, 29, 26, 92, 95, 98, 91, 56, 40, 63])
bins = [0, 25, 50, 75, 100]

plt.hist(a, bins)
plt.xticks([0, 25, 50, 75, 100]) # Can use the variable bins 
plt.ylabel('No of Students')
plt.xlabel('Grades')
plt.title('Exam Reuslts')
```

### VERY IMPORTANT FOR PRACTICAL 1
Image must be raveled (flattened)

```python
imbgr = cv2.imread('images/colourwheel.jpg')
imGray = cv2.cvtColor(imbgr, cv2.COLOR_BGR2GRAY)

plt.imshow(imGray, cmap='grey', vmin=0, vmax=255)
plt.show()

bins = 256
flattened_imGray = imGray.ravel()
print(flattened_imGray)

plt.hist(flattened_imGray, bins, range=[0,255])
plt.xlabel('Pixel Intensity')
plt.ylabel('No of pixels')
plt.title('Image Histogram')
plt.show()
```

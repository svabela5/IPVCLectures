```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Converting using openCV

```python
imBGR = cv2.imread('images/bird2.jpg')

imGray = cv2.cvtColor(imBGR, cv2.COLOR_BGR2GRAY)
plt.imshow(imGray, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Extracting the brightness channel from hsv

```python
imBGR = cv2.imread('images/bird2.jpg')

imHSV = cv2.cvtColor(imBGR, cv2.COLOR_BGR2HSV)
imGray = imHSV[:,:,2]

plt.imshow(imGray, cmap='gray', vmin=0, vmax=255)
plt.show()
```

### Using the formula

```python
imBGR = cv2.imread('images/bird2.jpg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)

Red_Channel = imRGB[:, :, 0]
Green_Channel = imRGB[:, :, 1]
Blue_Channel = imRGB[:, :, 2]

imGray = (0.298936 * Red_Channel + 
          0.587043 * Green_Channel + 
          0.114021 * Blue_Channel)

plt.imshow(imGray, cmap='gray', vmin=0, vmax=255)
plt.show()
```
